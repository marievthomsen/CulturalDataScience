## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")

```

```python
# Combines two tables from wiki page
df1 = rawdata[0] # table 1
df2 = rawdata[3] #table 4

# Adds column with type of single
df1['Type'] = 'Physical'
df2['Type'] = 'Digital'

# Concatenate both dataframes into one
df = pd.concat([df1, df2], ignore_index=True)

df
```

```python
# Output single for 2003
# Ensure the 'Released' column is numerical for year-based operations
df['Released'] = pd.to_numeric(df['Released'], errors='coerce')

# Sort the DataFrame by release year for easier processing
df = df.sort_values(by='Released').reset_index(drop=True)

# Target years
birth_year = 2003
year_15 = birth_year + 15  # 2018

# Process for 2003: Find the entry for 2003 or the closest year after 2003
if birth_year in df['Released'].values:
    result_2003 = df[df['Released'] == birth_year].iloc[0]
else:
    result_2003 = df[df['Released'] > birth_year].iloc[0]

artist_2003 = result_2003['Artist']
single_2003 = result_2003['Single']
year_2003 = result_2003['Released']

print(f"Artist and Single for closest year to 2003: {artist_2003}, {single_2003} (Year: {year_2003})")

# Process for 2018: Find the entry for 2018 or the closest year after 2018
if year_15 in df['Released'].values:
    result_2018 = df[df['Released'] == year_15].iloc[0]
else:
    result_2018 = df[df['Released'] > year_15].iloc[0]

artist_2018 = result_2018['Artist']
single_2018 = result_2018['Single']
year_2018 = result_2018['Released']

print(f"Artist and Single for closest year to 2018: {artist_2018}, {single_2018} (Year: {year_2018})")

```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")

df1 = rawdata[0]
df2 = rawdata[1]
df3 = rawdata[2]
df4 = rawdata[3]
df5 = rawdata[4]
df6 = rawdata[5]
df7 = rawdata[6]
df8 = rawdata[7]

df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)
```

```python
df
```

```python
# Finding the most common year
## Extract the last four digits in launch date and create a new column
df['Year'] = df['Launch date'].str[-4:]

## Count occurrences
counts = df['Year'].value_counts()

## Find the most common year and how many times it occurs
most_common = counts.idxmax()
most_common_count = counts.max()

# Output results
print(f"Year with most launches: {most_common} (occurs {most_common_count} times)")
```

```python
# Convert Launch date to datetime format
df['Launch date'] = pd.to_datetime(df['Launch date'], format='%d %B %Y')

# Extract the month
df['Month'] = df['Launch date'].dt.month_name()

# Count occurrences
counts_month = df['Month'].value_counts()

# Find the most common month
most_common_month = counts_month.idxmax()
most_common_count_month = counts_month.max()

# Output results
print(f"Month with most launches: {most_common_month} (occurs {most_common_count_month} times)")
```

```python
# Step 4: Sort the counts in descending order and reset the index
ranked_months = counts_month.sort_values(ascending=False).reset_index()

# Rename columns for clarity
ranked_months.columns = ['Month', 'Launch Count']

# Output the ranked DataFrame
print(ranked_months)
```

```python
rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_comic_book_supervillain_debuts")
```

```python
df1 = rawdata[0]
df2 = rawdata[1]
df3 = rawdata[2]
df4 = rawdata[3]
df5 = rawdata[4]
df6 = rawdata[5]
df7 = rawdata[6]
df8 = rawdata[7]
df9 = rawdata[8]

df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9], ignore_index=True)

# Delete the first 5 rows and first 2 columns
df = df.drop(index=df.index[:5]).reset_index(drop=True)
df = df.drop(df.columns[[0, 1]], axis=1)


df
```

```python
df['Year'] = df['Year Debuted'].str[0:4]

# Count occurrences
count_year = df['Year'].value_counts()

# Find the most common year
most_common_year = count_year.idxmax()
most_common_count_year = count_year.max()

# Sort the counts in descending order and reset the index
ranked_years = count_year.sort_values(ascending=False).reset_index()

# Rename columns
ranked_years.columns = ['Year', 'Launch Count']

# Output the ranked DataFrame
print(ranked_years)
```

```python
# Step 1: Group by Company and count occurrences
company_counts = df['Company'].value_counts()

# Step 2: Convert to DataFrame
ranked_companies = company_counts.reset_index()

# Step 3: Rename the columns for clarity
ranked_companies.columns = ['Company', 'Number of Supervillains']

# Step 4: Sort the DataFrame by number of supervillains in descending order
ranked_companies = ranked_companies.sort_values(by='Number of Supervillains', ascending=False)

# Output the ranked DataFrame
print(ranked_companies)
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
```

```python
# Starter code 2...

# Make a new column using the dataframe indices as particpant numbers

df.index.name = 'Participant_id'
df = df.reset_index()

#df.reset_index(inplace = True)
# NOTE: This line does exactly the same thing as the line above:
# it replaces the original df with a new df with the updated index. That's what
# "inplace = True" means. Or, you can just assign the dataframe with the updated index
# to a new dataframe with the same name as the old dataframe, which is what I did above.
# The end result is the same.

df
```

```python
long_df = pd.melt(df, id_vars=['Participant_id', 'Year'], 
                  value_vars=['Reading_NoInt', 'Naming_Int', 'Naming_NoInt', 'Reading_Int'],
                  var_name='Task', value_name='RT')
long_df
```
